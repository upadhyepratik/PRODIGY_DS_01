**Problem Statement**:

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

**Sample Dataset :-**

https://data.worldbank.org/indicator/SP.POP.TOTL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from CSV file
df = pd.read_csv('/content/drive/MyDrive/PRODIGY_DATA_SCIENCE_INTERNSHIP/API_SP.POP.TOTL_DS2_en_excel_v2_6011291.csv')

# Displays the dataset used
df
```

```python
parameters = df.describe()
print(parameters)              #this generates statistics sorted by years from 1960 to 2022
```

```python
# Drop unuseful columns
df.drop(columns=['Country Code','Indicator Name','Indicator Code'],inplace=True)
df
```

# **Data visualization**

```python
# sorting data based on Highest population in 2022
population_top = df.sort_values(by=['2022'],ascending=False).head(10)
population_top
```

```python
# Bar Plot
plt.figure(figsize=(18,9))
plt.rcParams['axes.facecolor']=="black"

plt.subplot(2,1,1)
plot=sns.barplot(x='1960',y="Country Name",data=population_top,palette="viridis")
plt.title("Highly Populated Country in the year 1960")
plt.xlabel("Population")
plt.ylabel("Country")

plt.subplot(2,1,2)
plot=sns.barplot(x='2022',y="Country Name",data=population_top,palette="viridis")
plt.title("Highly Populated Country in the year 2022")
plt.xlabel("Population")
plt.ylabel("Country")

plt.tight_layout()  #automatically adjust the padding between and around subplots
```

```python
# sorting data based on Least population in 2022
population_low = df.sort_values(by=['2022'],ascending=True).head(10)
population_low
```

```python
# Bar Plot
plt.figure(figsize=(18,9))
plt.rcParams['axes.facecolor']=="black"

plt.subplot(2,1,1)
plot=sns.barplot(x='1960',y="Country Name",data=population_low,palette="viridis")
plt.title("Least Populated Country in the year 1960")
plt.xlabel("Population")
plt.ylabel("Country")

plt.subplot(2,1,2)
plot=sns.barplot(x='2022',y="Country Name",data=population_low,palette="viridis")
plt.title("Least Populated Country in the year 2022")
plt.xlabel("Population")
plt.ylabel("Country")

plt.tight_layout()  #automatically adjust the padding between and around subplots
```
